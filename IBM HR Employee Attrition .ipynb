```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Attrition'].value_counts(normalize=True)*100
sns.countplot(data=df, x='Attrition')
plt.title("Employee Attrition Split")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.xticks(rotation=45)
plt.title("Attrition by Job Role")
plt.show()
```

```python
sns.boxplot(data=df, x='Attrition', y='DistanceFromHome')
plt.title("Distance From Home vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Education', y='MonthlyIncome', hue='Attrition')
plt.title("Avg Monthly Income by Education & Attrition")
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("cleaned_HR-Employee-Attrition.csv", index=False)
print("\nâœ… Cleaned file saved as 'cleaned_HR-Employee-Attrition.csv'")
```
